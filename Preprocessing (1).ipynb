```python
from google.colab import drive
import os
from PIL import Image

# 1. Hubungkan Google Colab ke Google Drive
drive.mount('/content/drive')

# 2. Atur Lokasi Folder (KHUSUS KELAS 1000 TERANG)
# Asumsi: folder DATASET_DUIT ada di halaman depan Google Drive kamu
folder_asal = '/content/drive/MyDrive/DATASET_DUIT/100000_kertas/100000_terang1'

# Kita buat folder penampungan baru bernama DATASET_SIAP agar file asli aman
folder_tujuan = '/content/drive/MyDrive/DATASET_DUIT/100000_kertas/100000_terang'

# Perintah ini akan otomatis membuat folder tujuan di Google Drive jika belum ada
if not os.path.exists(folder_tujuan):
    os.makedirs(folder_tujuan)

# 3. Proses Resize ke 224x224 dan Kompresi (Sesuai instruksi dosen)
ukuran_target = (224, 224)
foto_berhasil = 0

print(f"Mulai memproses foto di folder:\n{folder_asal}...\n")

# Cek semua file di dalam folder asal
try:
    for nama_file in os.listdir(folder_asal):
        # Hanya memproses file gambar
        if nama_file.lower().endswith(('.png', '.jpg', '.jpeg')):
            path_asal = os.path.join(folder_asal, nama_file)

            try:
                # Buka gambar dan ubah ke format standar RGB (wajib untuk standarisasi format)
                img = Image.open(path_asal).convert('RGB')

                # Resize gambar jadi 224x224
                img_resized = img.resize(ukuran_target, Image.Resampling.LANCZOS)

                # Pastikan format akhirnya seragam menjadi .jpg (sesuai instruksi dosen)
                nama_file_baru = os.path.splitext(nama_file)[0] + '.jpg'
                path_tujuan_baru = os.path.join(folder_tujuan, nama_file_baru)

                # Simpan dengan kualitas 85 (dari 10MB bisa jadi sekitar 30-50 KB saja!)
                img_resized.save(path_tujuan_baru, format='JPEG', quality=85)
                foto_berhasil += 1

                print(f"✅ Berhasil kompres: {nama_file_baru}")

            except Exception as e:
                print(f"❌ Gagal memproses file {nama_file}: {e}")

    print(f"\n🎉 MANTAP! Total {foto_berhasil} foto '100000 Terang' sukses dikompres & di-resize ke 224x224.")

except FileNotFoundError:
    print(f"❌ ERROR: Folder {folder_asal} tidak ditemukan.")
    print("Pastikan folder DATASET_DUIT benar-benar ada di halaman depan (root) Google Drive kamu, bukan di dalam folder lain.")
```

```python
from PIL import Image, ImageEnhance
import os

# 1. Pastikan Google Drive masih terhubung (jika belum, jalankan drive.mount lagi)
try:
    if not os.path.exists('/content/drive'):
        from google.colab import drive
        drive.mount('/content/drive')
except:
    pass

# 2. Atur Lokasi Folder (KHUSUS KELAS 1000)
# Kita ambil gambar yang sudah di-resize (DATASET_SIAP)
folder_sumber = '/content/drive/MyDrive/DATASET_DUIT/100000_kertas/100000_terang'

# Kita buat folder tujuan baru di Drive (sejajar dengan folder terang)
folder_gelap = '/content/drive/MyDrive/DATASET_DUIT/100000_kertas/100000_gelap'

if not os.path.exists(folder_gelap):
    os.makedirs(folder_gelap)

# 3. Kunci Rahasia: Faktor Kegelapan
# Nilai 1.0 artinya gambar asli.
# Nilai di bawah 1.0 artinya semakin gelap.
# Nilai 0.0 artinya hitam total.
# Untuk efek "ruangan gelap tapi objek masih terlihat", nilai 0.2 - 0.3 adalah yang paling pas.
faktor_kegelapan = 0.2

foto_gelap_berhasil = 0

print(f"Mulai menyulap foto di folder:\n{folder_sumber}\nmenjadi gelap...\n")

# Cek semua file di folder sumber
try:
    for nama_file in os.listdir(folder_sumber):
        if nama_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            path_asal = os.path.join(folder_sumber, nama_file)

            try:
                # Buka gambar asli
                img = Image.open(path_asal)

                # Masuk ke mode augmentasi kecerahan
                enhancer = ImageEnhance.Brightness(img)

                # Terapkan faktor kegelapan
                img_gelap = enhancer.enhance(faktor_kegelapan)

                # Simpan dengan nama baru, contoh: 'gelap_foto1.jpg'
                nama_file_baru = 'gelap_' + os.path.splitext(nama_file)[0] + '.jpg'
                path_tujuan_baru = os.path.join(folder_gelap, nama_file_baru)

                # Simpan sebagai JPG
                img_gelap.save(path_tujuan_baru, format='JPEG', quality=90)

                foto_gelap_berhasil += 1
                print(f"🌑 Berhasil menggelapkan: {nama_file_baru}")

            except Exception as e:
                print(f"❌ Gagal menggelapkan {nama_file}: {e}")

    print(f"\n🎉 SUKSES BESAR! Total {foto_gelap_berhasil} foto '100000_gelap' baru berhasil dibuat.")
    print(f"Cek di Drive kamu, folder: DATASET_DUIT/100000_kertas/100000_gelap")

except FileNotFoundError:
    print(f"❌ ERROR: Folder sumber {folder_sumber} tidak ditemukan.")
    print("Pastikan kamu sudah menjalankan kode kompresi kelas 1000 sebelumnya.")
```

```python
import os
import cv2
import numpy as np
from google.colab import drive

# 1. Hubungkan ke Google Drive
drive.mount('/content/drive')

# GANTI INI: Pastikan mengarah ke folder yang fotonya SUDAH KECIL (224x224)
path_dataset = '/content/drive/MyDrive/DATASET_DUIT'

data_gambar = []
label_kelas = []

# Daftar kelas
daftar_kelas = ['1000', '2000', '5000', '10000', '20000', '50000', '75000', '100000']
# Daftar kondisi (anak folder)
daftar_kondisi = ['terang', 'gelap']

print("Mulai masuk ke dalam folder dan memuat gambar...")

# 3. Looping untuk membaca semua gambar di setiap anak folder
for kelas in daftar_kelas:
    for kondisi in daftar_kondisi:
        # Ini akan membuat path seperti: DATASET_SIAP/1000_kertas/1000_terang
        nama_folder_anak = f"{kelas}_{kondisi}"
        folder_target = os.path.join(path_dataset, f"{kelas}_kertas", nama_folder_anak)

        # Cek apakah foldernya ada
        if not os.path.exists(folder_target):
            continue

        for nama_file in os.listdir(folder_target):
            if nama_file.lower().endswith(('.jpg', '.jpeg', '.png')):
                path_file = os.path.join(folder_target, nama_file)

                # Baca gambar
                img = cv2.imread(path_file)

                # Jika gambar berhasil dibaca
                if img is not None:
                    # Standarisasi Channel RGB
                    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

                    # STANDARISASI UKURAN: Jaga-jaga kalau ada foto yang belum 224x224
                    img = cv2.resize(img, (224, 224))

                    # Masukkan ke wadah
                    data_gambar.append(img)
                    label_kelas.append(kelas) # Label yang disimpan cukup nominalnya saja ('1000', dll)

# 4. NORMALISASI PIXEL (0-255 menjadi 0-1)
print("Memulai proses normalisasi (ini mungkin memakan waktu beberapa detik)...")
X = np.array(data_gambar, dtype='float32')
y = np.array(label_kelas)

# Pembagian untuk normalisasi
X_normalisasi = X / 255.0

print(f"\n✅ MANTAP! Semua instruksi selesai.")
print(f"📊 Total gambar berhasil dimuat : {len(X)}")
print(f"📊 Rentang nilai pixel sebelum normalisasi : {np.min(X)} sampai {np.max(X)}")
print(f"📊 Rentang nilai pixel sesudah normalisasi : {np.min(X_normalisasi)} sampai {np.max(X_normalisasi)}")
```

```python
from sklearn.model_selection import train_test_split
import collections

# Membagi data: 80% untuk Training, 20% untuk Testing
# random_state=42 agar hasil acakannya konsisten (tidak berubah-ubah kalau di-run ulang)
X_train, X_test, y_train, y_test = train_test_split(
    X_normalisasi,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y # stratify memastikan pembagian per kelasnya seimbang (balance)
)

print("✅ PROSES SPLIT DATA SELESAI!")
print("-" * 30)
print(f"Total seluruh data : {len(X_normalisasi)} gambar")
print(f"Data untuk Belajar (Training) : {len(X_train)} gambar (80%)")
print(f"Data untuk Ujian (Testing)  : {len(X_test)} gambar (20%)")

print("\n📊 Detail Jumlah Data Ujian per Kelas (Testing):")
# Menghitung jumlah masing-masing kelas di data testing
hitung_test = collections.Counter(y_test)
for kelas, jumlah in hitung_test.items():
    print(f"Kelas Rp {kelas}: {jumlah} gambar")
```
